```python
import re
g_loss=[]
d_loss=[]
with open("kuber_gan_error.txt","r") as f:
    for line in f:
        if "Epoch" in line:
            pattern = r"Epoch \[(\d+)/100\], G Loss: ([\d.]+), D Loss: ([\d.]+)"
            # Searching the string for matches
            match = re.search(pattern, line)
            if match:
                  # The epoch number
                g_loss_value = match.group(2)  # The G loss value
                d_loss_value = match.group(3)  # The D loss value

                g_loss.append(float(g_loss_value))
                d_loss.append(float(d_loss_value))
            
```

```python
for i in g_loss:
    print(type(i))
```

```python
import matplotlib.pyplot as plt
epoch=range(1,101)
plt.plot(epoch,g_loss)
plt.plot(epoch,d_loss)
```
